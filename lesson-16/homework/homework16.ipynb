```python
class NamedArray(np.ndarray):
    def __new__(cls, input_array, name="Unnamed"):
        obj = np.asarray(input_array).view(cls)
        obj.name = name
        return obj

    def __array_finalize__(self, obj):
        if obj is None: return
        self.name = getattr(obj, 'name', "Unnamed")

# Example usage
data = np.array([[1, 2, 3], [4, 5, 6]])
named_array = NamedArray(data, name="MyArray")
print(named_array)
print("Name:", named_array.name)
```

```python
def extract_centered_subarray(arr, center, shape, fill=0):
    arr = np.asarray(arr)
    out = np.full(shape, fill, dtype=arr.dtype)
    
    # Calculate slices
    in_slices = []
    out_slices = []
    for c, s, maxdim in zip(center, shape, arr.shape):
        start = c - s // 2
        end = start + s

        in_start = max(start, 0)
        in_end = min(end, maxdim)
        out_start = max(-start, 0)
        out_end = out_start + (in_end - in_start)

        in_slices.append(slice(in_start, in_end))
        out_slices.append(slice(out_start, out_end))

    out[tuple(out_slices)] = arr[tuple(in_slices)]
    return out
A = np.arange(25).reshape(5, 5)
print(extract_centered_subarray(A, center=(0, 0), shape=(3, 3), fill=-1))
```

```python
class SymmetricArray(np.ndarray):
    def __getitem__(self, idx):
        i, j = idx
        return super().__getitem__((min(i,j), max(i,j)))

    def __setitem__(self, idx, value):
        i, j = idx
        super().__setitem__((min(i,j), max(i,j)), value)

# Create one
Z = SymmetricArray((5, 5), buffer=np.zeros((5,5)))
Z[2, 3] = 42
print(Z[3, 2])  # 42
```

```python
def game_of_life_step(Z):
    neighbors = sum(np.roll(np.roll(Z, i, 0), j, 1)
                    for i in (-1, 0, 1) for j in (-1, 0, 1)
                    if (i != 0 or j != 0))
    return (neighbors == 3) | ((Z == 1) & (neighbors == 2)).astype(int)

# Example
Z = np.random.randint(0, 2, (10, 10))
Z = game_of_life_step(Z)
```
