```python
bins = [0, 200_000, 400_000, 600_000, 800_000, 1000_000,
        1200_000, 1400_000, 1600_000, 1800_000, np.inf]

population['salary_band'] = pd.cut(population['salary'], bins=bins, labels=psa['Salary Band'])
```

```python
population
psa = population.groupby('salary_band', observed=False)['salary'].agg(['count', 'mean', 'median', 'sum'])
```

```python
psa['percentage'] = (psa['count'] / len(population))*100
```

```python
state = population.groupby('state', observed=False)['salary'].agg(['count', 'mean', 'median', 'sum'])
state['percentage'] = (state['count'] / len(population))*100
state
```
