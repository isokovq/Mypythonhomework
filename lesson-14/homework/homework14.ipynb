```python
import json
import os
from uuid import uuid4

# Sample initial books.json structure if file doesn't exist
INITIAL_BOOKS = {
    "books": [
        {
            "id": str(uuid4()),
            "title": "The Great Gatsby",
            "author": "F. Scott Fitzgerald",
            "year": 1925,
            "isbn": "978-0743273565"
        },
        {
            "id": str(uuid4()),
            "title": "1984",
            "author": "George Orwell",
            "year": 1949,
            "isbn": "978-0451524935"
        }
    ]
}

# File path for books.json
FILE_PATH = "books.json"

def load_books():
    """Load books from books.json or create a new file if it doesn't exist."""
    if not os.path.exists(FILE_PATH):
        with open(FILE_PATH, 'w') as f:
            json.dump(INITIAL_BOOKS, f, indent=4)
        return INITIAL_BOOKS["books"]
    try:
        with open(FILE_PATH, 'r') as f:
            data = json.load(f)
        return data["books"]
    except json.JSONDecodeError:
        print("Error: Invalid JSON format in books.json. Initializing with default data.")
        with open(FILE_PATH, 'w') as f:
            json.dump(INITIAL_BOOKS, f, indent=4)
        return INITIAL_BOOKS["books"]

def save_books(books):
    """Save books to books.json."""
    with open(FILE_PATH, 'w') as f:
        json.dump({"books": books}, f, indent=4)

def add_book():
    """Add a new book to the JSON file."""
    books = load_books()
    title = input("Enter book title: ").strip()
    author = input("Enter author name: ").strip()
    year = input("Enter publication year (e.g., 2023): ").strip()
    isbn = input("Enter ISBN: ").strip()

    try:
        year = int(year)
        if year < 0:
            raise ValueError("Year cannot be negative.")
    except ValueError:
        print("Error: Invalid year format. Book not added.")
        return

    new_book = {
        "id": str(uuid4()),
        "title": title,
        "author": author,
        "year": year,
        "isbn": isbn
    }
    books.append(new_book)
    save_books(books)
    print(f"Book '{title}' added successfully!")

def update_book():
    """Update an existing book's information."""
    books = load_books()
    book_id = input("Enter the book ID to update: ").strip()
    book = next((b for b in books if b["id"] == book_id), None)

    if not book:
        print("Error: Book ID not found.")
        return

    print(f"Current book details: {book}")
    title = input(f"Enter new title (or press Enter to keep '{book['title']}'): ").strip()
    author = input(f"Enter new author (or press Enter to keep '{book['author']}'): ").strip()
    year = input(f"Enter new year (or press Enter to keep {book['year']}): ").strip()
    isbn = input(f"Enter new ISBN (or press Enter to keep '{book['isbn']}'): ").strip()

    if title:
        book["title"] = title
    if author:
        book["author"] = author
    if year:
        try:
            year = int(year)
            if year < 0:
                raise ValueError("Year cannot be negative.")
            book["year"] = year
        except ValueError:
            print("Error: Invalid year format. Year not updated.")
    if isbn:
        book["isbn"] = isbn

    save_books(books)
    print(f"Book ID {book_id} updated successfully!")

def delete_book():
    """Delete a book from the JSON file."""
    books = load_books()
    book_id = input("Enter the book ID to delete: ").strip()
    book = next((b for b in books if b["id"] == book_id), None)

    if not book:
        print("Error: Book ID not found.")
        return

    print(f"Book to delete: {book}")
    confirm = input("Confirm deletion (y/n): ").strip().lower()
    if confirm == 'y':
        books[:] = [b for b in books if b["id"] != book_id]
        save_books(books)
        print(f"Book ID {book_id} deleted successfully!")
    else:
        print("Deletion cancelled.")

def display_books():
    """Display all books in the JSON file."""
    books = load_books()
    if not books:
        print("No books found.")
        return
    for book in books:
        print(f"ID: {book['id']}")
        print(f"Title: {book['title']}")
        print(f"Author: {book['author']}")
        print(f"Year: {book['year']}")
        print(f"ISBN: {book['isbn']}")
        print("-" * 40)

def main():
    """Main program loop."""
    while True:
        print("\nBook Management System")
        print("1. Add a new book")
        print("2. Update a book")
        print("3. Delete a book")
        print("4. Display all books")
        print("5. Exit")
        choice = input("Enter your choice (1-5): ").strip()

        if choice == "1":
            add_book()
        elif choice == "2":
            update_book()
        elif choice == "3":
            delete_book()
        elif choice == "4":
            display_books()
        elif choice == "5":
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
