```python
import threading
import time

lst = []

def prime_checker(a, b):
    for i in range(a, b + 1):
        if i > 1:
            for j in range(2, int(i**0.5) + 1):
                if (i % j) == 0:
                    break
            else:
                lst.append(i)
                print(f"Prime found: {i}")
        time.sleep(0.1)  # Simulate some delay

# Define threads
thread1 = threading.Thread(target=prime_checker, args=(1, 100))
thread2 = threading.Thread(target=prime_checker, args=(101, 200))
thread3 = threading.Thread(target=prime_checker, args=(201, 300))
thread4 = threading.Thread(target=prime_checker, args=(301, 400))

# Start threads
thread1.start()
thread2.start()
thread3.start()
thread4.start()

# Wait for threads to complete
thread1.join()
thread2.join()
thread3.join()
thread4.join()

# Display results neatly
print("\nAll threads have completed execution.")
print(f"Total primes found: {len(lst)}")
print(f"Primes: {sorted(lst)}")
```

```python
with open("bigtext.txt", mode ="r", encoding = "utf-8") as file:
    text = file.read()
    text = text.split()
def word_counter(num):
    for word in text[:num]:
        print(f"the word '{word}' appeared {str(text[:num].count(word))} times")
    time.sleep(0.65)  # Simulate some delay
thread1 = threading.Thread(target=word_counter, args=(100,))
thread2 = threading.Thread(target=word_counter, args=(200,))
thread3 = threading.Thread(target=word_counter, args=(300,))

thread1.start()
thread2.start()
thread3.start()

thread1.join()
thread2.join()
thread3.join()
print("\nAll threads have completed execution.")
```
