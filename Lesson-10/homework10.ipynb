```python
class Task:
    def __init__ (self, task_title, description, due_date, status=False):
        self.task_title = task_title
        self.description = description
        self.due_date = due_date
        self.status = status
class ToDoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, task_title, description, due_date):
        new_task = Task(task_title, description, due_date)
        self.tasks.append(new_task)

    def remove_task(self, task_title):
        for task in self.tasks:
            if task.task_title == task_title:
                self.tasks.remove(task)
                break

    def mark_as_done(self, task_title):
        for task in self.tasks:
            if task.task_title == task_title:
                task.status = True
                break

    def display_tasks(self):
        for task in self.tasks:
            status = "Done" if task.status else "Not Done"
            print(f"Title: {task.task_title}, Description: {task.description}, Due Date: {task.due_date}, Status: {status}")
    
    def display_not_done_tasks(self):
        for task in self.tasks:
            if not task.status:
                print(f"Title: {task.task_title}, Description: {task.description}, Due Date: {task.due_date}, Status: Not Done")

def main():
    todo_list = ToDoList()
    while True:
        print("\nTo-Do List Menu:")
        print("1. Add Task")
        print("2. Remove Task")
        print("3. Mark Task as Done")
        print("4. Display All Tasks")
        print("5. Display Not Done Tasks")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            task_title = input("Enter task title: ")
            description = input("Enter task description: ")
            due_date = input("Enter task due date (YYYY-MM-DD): ")
            todo_list.add_task(task_title, description, due_date)
            print(f"Task '{task_title}' added.")
        elif choice == '2':
            task_title = input("Enter task title to remove: ")
            todo_list.remove_task(task_title)
            print(f"Task '{task_title}' removed.")
        elif choice == '3':
            task_title = input("Enter task title to mark as done: ")
            todo_list.mark_as_done(task_title)
            print(f"Task '{task_title}' marked as done.")
        elif choice == '4':
            todo_list.display_tasks()
        elif choice == '5':
            todo_list.display_not_done_tasks()
        elif choice == '6':
            break
        else:
            print("Invalid choice. Please try again.")
            
```

```python
main()
```

```python
class Post:
    def __init__(self, title, content, author):
        self.title = title
        self.content = content
        self.author = author
class Blog:
    def __init__ (self):
        self.posts = []
    def add_post(self, title, content, author):
        new_post = Post(title, content, author)
        self.posts.append(new_post)
    def list_posts(self):
        for post in self.posts:
            print(f"Title: {post.title}, Author: {post.author}")
    def display_by_author(self, author):
        for post in self.posts:
            if post.author == author:
                print(f"Title: {post.title}, Content: {post.content}")
    def delete_post(self, title):
        for post in self.posts:
            if post.title == title:
                self.posts.remove(post)
                break
    def edit_post(self, title, new_title, new_content):
        for post in self.posts:
            if post.title == title:
                post.title = new_title
                post.content = new_content
                break            
    def display_latest_posts(self, n=5):
        latest_posts = sorted(self.posts, key=lambda x: x.title, reverse=True)[:n]
        for post in latest_posts:
            print(f"Title: {post.title}, Author: {post.author}")            

def main():
    blog = Blog()
    while True:
        print("\nBlog Menu:")
        print("1. Add Post")
        print("2. List Posts")
        print("3. Display Posts by Author")
        print("4. Delete Post")
        print("5. Edit Post")
        print("6. Display Latest Posts")
        print("7. Exit")

        choice = input("Enter your choice: ")
        if choice == '1':
            title = input("Enter post title: ")
            content = input("Enter post content: ")
            author = input("Enter post author: ")
            blog.add_post(title, content, author)
            print(f"Post '{title}' added.")
        elif choice == '2':
            blog.list_posts()
        elif choice == '3':
            author = input("Enter author name to filter posts: ")
            blog.display_by_author(author)
        elif choice == '4':
            title = input("Enter post title to delete: ")
            blog.delete_post(title)
            print(f"Post '{title}' deleted.")
        elif choice == '5':
            title = input("Enter post title to edit: ")
            new_title = input("Enter new post title: ")
            new_content = input("Enter new post content: ")
            blog.edit_post(title, new_title, new_content)
            print(f"Post '{title}' edited.")
        elif choice == '6':
            n = int(input("Enter number of latest posts to display: "))
            blog.display_latest_posts(n)
        elif choice == '7':
            break
        else:
            print("Invalid choice. Please try again.")

```

```python
main()
```
